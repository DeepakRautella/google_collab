**API**

API stands for Application Programming Interface. In the context of APIs, the word Application refers to any software with a distinct function. Interface can be thought of as a contract of service between two applications. This contract defines how the two communicate with each other using requests and responses.

you need server side api to load files

```python
import pandas as pd
import requests
```

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<apikey>>&language=en-US&page=1')
```

```python
response.json()['results']
```

```python
pd.DataFrame(response.json()['results']).head()
```

```python
pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df=pd.DataFrame()
```

```python
for i in range(1,499):
  response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<apikey>>&language=en-US&page={i}').format(i))
  temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df=df.append(temp_df,ignore_index=True)
```
